- ### 7) Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30).

We have already converted the *tpep_pickup_datetime* and *tpep_pickup_datetime* variables to *datetime*, so we can pick both time and date with the *.dt* functions and save them in new columns. Specifically we're going to need the *pickup_time* variable for later. 

```python
data["pickup_date"]=data['tpep_pickup_datetime'].dt.date #saving the pickup date in a new column.
data["pickup_time"]=data['tpep_pickup_datetime'].dt.time #saving the pickup time in a new column.

data["dropoff_date"]=data['tpep_dropoff_datetime'].dt.date #doing the same for dropoff date and time.
data["dropoff_time"]=data['tpep_dropoff_datetime'].dt.time

data.head(1) #for the first trip, pickup_date is 2020-01-01, pickup_time is 00:28:15, dropoff_date is 2020-01-01, dropoff_time is 00:33:03.
```

Now we define the function *interval* that, for each *pickup_time*, returns the 30-minute interval it's in. To make it work then, we have to convert *pickup_time* in the *datetime H:M:S* format.

```python
def interval(time):
    m=time.minute #minutes.
    h=time.hour #hours.
    
    if m<30: #case where the time falls into the first half of the hour.
        return f"{h:02d}:00-{h:02d}:30" #writing the interval.       
    else: #case where the time falls into the second half of the hour.
        next_h=(h+1)%24 #getting the next hour, (remembering that after 23 comes 0).
        return f"{h:02d}:30-{next_h:02d}:00" #writing the interval.

data['pickup_time']=pd.to_datetime(data['pickup_time'], format='%H:%M:%S').dt.time #converting pickup_time in the datetime H:M:S format.
data['interval']=data['pickup_time'].apply(interval) #applying the function on the dataframe.
data.head(1) #the first trip falls into the 00:00 - 00:30 interval.
```

- ### 17) Build chains of trips.
#### Two trips are consecutive in a chain if:
- (a) they have the same *VendorID*.
- (b) the pickup location of the second trip is also the dropoff location of the first trip.
- (c) the pickup time of the second trip is after the dropoff time of the first trip.
- (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

Since obtaining every possible chain would mean using a recursive algorithm (which has complexity **O(n!)** and so is computationally impossible to execute), we decided to change our approach; we won't try to get every possible chain, but rather for each trip we'll add the column *chains_to* to show every possible other trip it can be chained to.

First, we'll try and clean the dataset, removing all trips with *trip_distance=0* or *total_amount<0*.

```python
data.shape[0] #the dataset is made of 6339567 rows.

data['trip_distance']=data['trip_distance'].astype(float) #converting trip_distance into float.
data=data[data['trip_distance']!=0] #removing trips with trip_distance=0.

data.shape[0] #now, the dataset is made of 6272091 rows.

data=data[data['total_amount']>=0] #removing trips where total_amount is negative (refunds).

data.shape[0] #now, the dataset is made of 6254985 rows.
```

Now we have the final dataset. We create the column *tripID* to enumerate the trips. 

```python
data["tripID"]=range(1,len(data)+1) #creating column tripID.
```

To ease the computation, we'll split the dataset in two smaller dataset on *VendorID* and then join them again later.

```python
V1=data[data['VendorID']==1] #dataset V1 is made of 2058650 rows.
V2=data[data['VendorID']==2] #dataset V2 is made of 4196335 rows.
```

This function, for each trip of the input dataset, returns all trips that can be chained to it into a list. The list is then saved into the column *chains_to*.

```python
def find_chains_to(dataset):
    
    dataset['chains_to']=[[] for _ in range(len(dataset))] #creating the column chains_to.
    
    dataset=dataset.sort_values(by='tpep_pickup_datetime') #sorting the dataset by tpep_pickup_datetime, so that we can consider only the rows under the selected row.

    grouped=dataset.groupby('PULocationID') #grouping the dataset by PULocationID.
    datasets={location: group.reset_index(drop=True) for location, group in grouped} #saving each group as a new dataset, to ease the computing later.
    
    chains_to_list=[] #creating a list to hold the results.

    for i in range(len(dataset)-1): #each row checks only the rows under it, since the dataset is sorted by tpep_pickup_datetime.
        row=dataset.iloc[i] #saving the selected row.

        if row["DOLocationID"] in datasets: #checking if there are other trips from the row's dropoff location.
            
            c1=datasets[row["DOLocationID"]]['tpep_pickup_datetime']>row['tpep_dropoff_datetime'] #checking that the first trip ends before the second trip starts.
            c2=(datasets[row["DOLocationID"]]['tpep_pickup_datetime']-row['tpep_dropoff_datetime']).dt.total_seconds()/60<=2  #checking that the second trip starts after at most 2 minutes.

            possible_chains= (datasets[row["DOLocationID"]][c1&c2]['tripID']).tolist() #saving all tripIDs of the trips that sastisfy both conditions.
            chains_to_list.append(possible_chains) #adding the results to the main list.
        else:
            chains_to_list.append([]) #if there are no other trips from the row's dropoff location, we simply add an empty list.

    chains_to_list.append([]) #we don't check for the last trip since there are no trips that can be chained to it by default, so we must add an empty list at the end.

    dataset['chains_to']=chains_to_list #putting the results into the chains_to column in the dataset.

    return dataset #returning the updated dataset.
```

We'll run this function for both datasets and then put them back together.

```python
V1=V1.reset_index(drop=True) #resetting the index.
V1=find_chains_to(V1)

V2=V2.reset_index(drop=True) #resetting the index.
V2=find_chains_to(V2)

data=pd.concat([V1,V2],ignore_index=True) #putting both datasets together again.
```

Last, we thought to define a function that, given a trip, would return one of the possible chains that can be built starting from it. It will be a recursive function, that breaks when it faces a trip with no possible chains to it.

```python
def build_chains(ID,data):
    print(ID) #printing current ID.
    if data[data["tripID"]==ID]["chains_to"].iloc[0]==[]: #checking if the current trip can be chained to other trips.
        return #if the current trip can't be chained, the function stops.
    else:
        build_chains(data[data["tripID"]==ID]["chains_to"].iloc[0][0],data) #repeating the process on the first of those trips that can be chained.
```

```python
build_chains(8387,data) #returning 8387,3540,7351.
```
